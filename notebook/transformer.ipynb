```python
import os
import re
import pandas as pd
import numpy as np
from sentence_transformers import SentenceTransformer, util
import matplotlib.pylab as plt
import matplotlib_fontja
os.chdir('../')
from src.feature import *
from src.model import *
```

```python
# データの読み込み
train = pd.read_csv("data/train.csv")
test = pd.read_csv("data/test.csv")
sample_submit = pd.read_csv("data/sample_submit.csv", header=None)

# Xとyの分割
target_col = '購入フラグ'
X_train_df = train.drop(columns=[target_col])
y_train_df = train[target_col]
X_test_df = test.copy()

# 学習・テストデータの結合
X_train_df.index = "train_" + X_train_df.index.astype(str)
X_test_df.index = "test_" + X_test_df.index.astype(str)
X_df = pd.concat([X_train_df, X_test_df], axis=0)
```

```python
# 埋め込みモデルの定義
model = SentenceTransformer(
    model_name_or_path='intfloat/multilingual-e5-base', 
    device='cpu',
    )

# 埋め込みモデルの学習
embeddings = model.encode(
    sentences=X_df["今後のDX展望"].to_list(),
    show_progress_bar=True,
    normalize_embeddings=False,
    )
embed_cols = [f"embed_{i}" for i in range(embeddings.shape[1])]
df_embed = pd.DataFrame(embeddings, columns=embed_cols, index=X_df.index)
df_embed.to_csv("data/今後のDX展望_embed.csv", index=True)
```
